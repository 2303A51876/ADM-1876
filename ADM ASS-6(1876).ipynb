```python
import pandas as pd
#Load dataset (replace with your file path or URL)
url = '//content/netflix_titles.csv'
data = pd.read_csv(url)
#generate statistical summary of the dataset
summary = data.describe()
#display the summary
print(summary)
```

```python
import matplotlib.pyplot as plt
plt.hist(data['country'].dropna(),bins=20,edgecolor='black',color='skyblue')
plt.title('country')
plt.xlabel('country')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar chart for 'Pclass' (Passenger class)
plt.figure(figsize=(6, 4))
data['type'].value_counts().plot(kind='bar', color='lightcoral')
plt.title('type')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
director_counts = data['director'].value_counts()
plt.bar(director_counts.index[:20], director_counts.values[:20], color='green')  # Limit to top 20 directors
plt.title('Top 20 Directors')
plt.xlabel('Director')
plt.ylabel('Number of Movies/Shows')
plt.xticks(rotation=90)

plt.show()
```

```python
import seaborn as sns
import pandas as pd

def extract_duration(duration_str):
    try:
        return int(duration_str.split()[0])
    except (ValueError, AttributeError):
        return None

# These lines should be outside the function definition
data['duration_numeric'] = data['duration'].apply(extract_duration)
numerical_features = ['duration_numeric']
sns.pairplot(data[numerical_features].dropna())
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('netflix_titles.csv')  # Make sure the filename matches
print(df.info())  # Display dataset information
print(df.head())  # Show first 5 rows
```

```python
print(df.columns)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.hist(df['release_year'].dropna(), bins=20, edgecolor='black', color='pink')
plt.title('Distribution of Netflix Titles by Release Year')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
print(df['rating'].unique())  # Display unique rating categories
```

```python
df['type'].value_counts().plot(kind='bar', color=['lightblue', 'orange'], edgecolor='black')
plt.title('Count of Movies vs. TV Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
df['rating'].value_counts().plot(kind='bar', color='lightblue', edgecolor='black')
plt.title('Distribution of Netflix Titles by Rating')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
print(df['rating'].unique())  # Show all unique rating categories
df['rating'].value_counts().plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Distribution of Netflix Titles by Rating')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='bar', color='purple', edgecolor='black')
plt.title('Top 10 Countries Producing Netflix Content')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
from collections import Counter
genres = ', '.join(df['listed_in'].dropna()).split(', ')
genre_counts = Counter(genres)
plt.figure(figsize=(12, 6))
plt.bar(genre_counts.keys(), genre_counts.values(), color='cyan', edgecolor='black')
plt.title('Most Popular Genres on Netflix')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```
