```python
import pandas as pd
data=pd.read_csv("/content/Titanic_dataset.csv")
print(data.head())
```

```python
#check for missing values
print(data.isnull().sum())
```

```python
#Impute missing values with mean (for numerical columns)
data['Age'].fillna(data['Age'].mean(), inplace=True)
#Alternatively, you can impute with the median
data['Age'].fillna(data['Age'].median(), inplace=True)
# Or, input with the mode (for categorical columns)
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
data['Cabin'].fillna(data['Cabin'].mode()[0], inplace=True)
#verify that missing values are handled
print(data.isnull().sum())
```

```python
#Check for missing values
print(data.isnull().sum())
```

```python
from sklearn.preprocessing import MinMaxScaler
#Initialize MinMaxScaler
scaler=MinMaxScaler()
#select columns to normalize(numerical columns)
columns_to_normalize=['Age','Fare']
#apply min-max normalization
data[columns_to_normalize]=scaler.fit_transform(data[columns_to_normalize])
#display the first few rows after normalization
data[columns_to_normalize].head()
```

```python
from sklearn.preprocessing import StandardScaler
#Initialize StandardScaler
scaler=StandardScaler()
#select columns to normalize(Standard columns)
columns_to_Standardize=['Age','Fare']
#apply Standard normalization
data[columns_to_Standardize]=scaler.fit_transform(data[columns_to_normalize])
#display the first few rows after normalization
data[columns_to_Standardize].head()
```
